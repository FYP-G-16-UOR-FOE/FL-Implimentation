```python
import matplotlib.pyplot as plt
import pickle
```

```python
with open('history/fedAvg_iid_accuracy_history.pkl', 'rb') as f:
    iid_accuracy_history = pickle.load(f)
    print("accuracy_history loaded from accuracy_history.pkl")

with open('history/fedAvg_non_iid_accuracy_history.pkl', 'rb') as f:
    non_iid_accuracy_history = pickle.load(f)
    print("accuracy_history loaded from accuracy_history.pkl")




# Plot both accuracy graphs
plt.figure(figsize=(12, 5))
plt.plot(iid_accuracy_history, label='IID', color='blue')
plt.plot(non_iid_accuracy_history, label='Non-IID', color='red')
plt.xlabel('Round')
plt.ylabel('Accuracy')
plt.title('FedAvg Algorithm Accuracy')
plt.legend()
plt.show()
```

```python
# Plot accuracy
plt.figure(figsize=(12, 5))
plt.plot(iid_accuracy_history, label='Accuracy', color='blue')
plt.xlabel('Round')
plt.ylabel('Accuracy')
plt.title('FedAvg Algorithm IID Accuracy')
plt.legend()
plt.show()

# Plot accuracy
plt.figure(figsize=(12, 5))
plt.plot(non_iid_accuracy_history, label='Accuracy', color='red')
plt.xlabel('Round')
plt.ylabel('Accuracy')
plt.title('FedAvg Algorithm non-IID Accuracy')
plt.legend()
plt.show()
```
